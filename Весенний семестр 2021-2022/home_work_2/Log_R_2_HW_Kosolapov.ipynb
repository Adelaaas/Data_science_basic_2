```python
train = train.drop('User ID', axis=1)
label_encoder = preprocessing.LabelEncoder() 
label_encoder.fit(train['Gender']) 
train['Gender'] = label_encoder.transform(train['Gender']) 

train.head(20)
```

```python
X = train.drop(['Purchased'], axis = 1)
y = train['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_st = scaler.fit_transform(X)
X_st
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_st, y, test_size=1/3, random_state=42)
```

```python
%%time

model = LogisticRegression()
model.fit(X_train, y_train)
answers_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,answers_pred))
```
