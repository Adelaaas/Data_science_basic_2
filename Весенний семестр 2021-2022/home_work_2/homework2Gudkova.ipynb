```python
X = df.drop(['class'], axis=1)
y = df['class']
X_df, X_test, y_df, y_test = train_test_split(X,y, test_size=0.3)
model = LogisticRegression()
model.fit(X_df, y_df)
answers_pred = model.predict(X_test)
```

```python
accuracy_score(y_test,answers_pred)
```

Точность - 95%
