```python
# 1. Считайте DataFrame
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("drug200.csv")
df
```

```python
# 2. Написать фукнцию, которая принимает DataFrame
# и возвращает кортеж (минимальный возраст, среднее значения возраста) для всех женщин
def func(x):
  y=x[x["Sex"]=="F"]
  return(y["Age"].min(),y["Age"].mean())
print(func(df))
```

```python
# 3. Построить график распределение по возрасту для каждого пола (м, ж)
# Графики расположены на одной координатной плоскости
# У графика должен быть заголовок, подписаны оси, а также легенда (т е графики должны быть подписаны)
m=df[df["Sex"]=="M"]
f=df[df["Sex"]=="F"]
sum=0
plt.title("Распределение по возрасту для каждого пола")
plt.xlabel("Возраст")
plt.ylabel("Количество")
plt.grid()
x=np.arange(df["Age"].min(),df["Age"].max()+1)
lst=np.zeros(df["Age"].max()+1-df["Age"].min())
lst1=lst.copy()
for i in m["Age"]:
  lst[i-df["Age"].min()]+=1
y1=lst.copy()
for i in f["Age"]:
  lst1[i-df["Age"].min()]+=1
y2=lst1.copy()
plt.plot(x,y1,label="Мужчины")
plt.plot(x,y2,label="Женщины")
plt.legend()
plt.show()
```
